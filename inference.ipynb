```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./statistics.log", "r") as f:
    lines = f.readlines()
content = eval(lines[-1].rstrip())

keys = ["00025(2,5)", "00100(2,10)", "00125(3,5)", "00256(4,4)", "00256(2,16)", "00625(4,5)", "00625(2,25)", 
                      "01000(3,10)", "02500(2,50)", "10000(2,100)", "10000(4,10)"]
x = range(11)
means = []
sems = []
for k in keys:
    means.append(content[k]["acc_mean"])
    sems.append(content[k]["acc_sem"])
```

```python
plt.figure(figsize=(20, 8))
plt.xlabel("input size")
plt.ylabel("test accuracy")
plt.xticks(x, keys)
plt.errorbar(x, means, sems, linestyle='None', marker='^')
plt.show()
```
